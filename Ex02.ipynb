```python

#Calcule a média, mediana e desvio padrão da primeira coluna (sepallength) do dataset Iris.

import numpy as np

data_url  = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = np.genfromtxt(data_url, delimiter=',', dtype='float', usecols=[0])

media = np.mean(iris)
mediana = np.median(iris)
desvio = np.std(iris)

print('Média = ' + str(media))
print('Mediana = ' + str(mediana))
print('Desvio Padrão = ' + str(desvio))
```

```python

#Normalize a coluna (sepallength) do dataset iris de tal forma que o valor mínimo seja zero (0) e o valor máximo seja 1.

import numpy as np

data_url  = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = np.genfromtxt(data_url, delimiter=',', dtype='float', usecols=[0])

minimo = min(iris)
maximo = max(iris)

for i in range(len(iris)):
    iris[i] = (iris[i] - minimo) / (maximo - minimo)

print(iris)
```

```python
#Filtre o vetor iris_2d do exercício anterior para que ele tenha: (terceira coluna (petallength) > 1.5) && (primeira coluna (sepallength) < 5.0)

import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float')
iris_2d[np.random.randint(150, size=20), np.random.randint(4, size=20)] = np.nan

terceira = iris_2d[:,2]
primeira = iris_2d[:,0]

for i in range(len(iris_2d)):
    if terceira[i] <= 1.5:
        terceira[i] = 0
    if primeira[i] >= 5:
        primeira[i] = 0

print('Terceira')
for i in range(len(iris_2d)):
    if terceira[i] != 0 and np.isnan(terceira[i]) == False:
        print(str(terceira[i]) + ', ')

print('Primeira')
for i in range(len(iris_2d)):
    if primeira[i] != 0 and np.isnan(primeira[i]) == False:
        print(str(primeira[i]) + ', ')
```

```python
#Substitua todas as ocorrências de nan (Not a Number) por zeros em um vetor NumPy.

import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
iris_2d[np.random.randint(150, size=20), np.random.randint(4, size=20)] = np.nan

for i in range(len(iris_2d)):
    for j in range(4):
        if np.isnan(iris_2d[i,j]).any() == True:
            iris_2d[i,j] = 0

print(iris_2d)

```

```python

#Mapeie terceira coluna (petallength) de tal forma que os valores numéricos sejam substituídos por valores textuais, 
#seguindo de acordo com a tabela abaixo:

#| Número               | Classificação                
#|:---------------------|:--------------|
#| Menor do que 3       | Pequena       |
#| entre 3 e 5          | Média         |
#| Maior ou Igual a 5   | Grande        |

import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[2])

iris_nova = []

for i in range(len(iris)):
    if iris[i] < 3:
        iris_nova.append('Pequena')
    elif iris[i] >= 3 and iris[i] < 5:
        iris_nova.append('Média')
    else:
        iris_nova.append('Grande')

print(iris_nova)
        
```

```python
#Encontre o valor mais frequente da terceira coluna (petallength) do dataset iris.

import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[2])

freq = 0
freq_posic = 0
list_iris = list(iris)

for i in range(len(iris)):
    if list_iris.count(list_iris[i]) > freq:
        freq = list_iris.count(iris[i])
        freq_posic = i
print(str(list_iris[freq_posic]) + ' é apresentado ' + str(freq) + ' vezes.')
```
